```python
import numpy as np
from scipy import misc
from matplotlib import pylab as plt
import matplotlib.cm as cm
%matplotlib inline
```

Load Data

```python
train_labels, train_data = [], []
for line in open('./faces/train.txt'):
    im = misc.imread(line.strip().split()[0])
    train_data.append(im.reshape(2500,))
    train_labels.append(line.strip().split()[1])
train_data, train_labels = np.array(train_data, dtype=float), np.array(train_labels, dtype=int)

test_labels, test_data = [], []
for line in open('./faces/test.txt'):
    im = misc.imread(line.strip().split()[0])
    test_data.append(im.reshape(2500,))
    test_labels.append(line.strip().split()[1])
test_data, test_labels = np.array(test_data, dtype=float), np.array(test_labels, dtype=int)

print train_data.shape, train_labels.shape, test_labels.shape, test_data.shape
plt.imshow(train_data[10, :].reshape(50,50), cmap = cm.Greys_r)
plt.show()
```

Average Face

```python
avg_face = np.sum(train_data, axis = 0) / train_data.shape[0]
plt.imshow(avg_face.reshape(50,50), cmap = cm.Greys_r)
plt.show()
```

SVD on training data

```python
#Creates just a figure and only one subplot
U,S,Vt = np.linalg.svd(train_data)
fig, axs = plt.subplots(2, 5)
fig.subplots_adjust( wspace=.7)
axs = axs.ravel()
for i in range(0,10):
    eigen_face = Vt[i,:]
    axs[i].imshow(eigen_face.reshape(50,50), cmap = cm.Greys_r)
plt.tight_layout()
plt.show()
```

error for Low Rank approximation using SVD

```python
error = np.ndarray(shape = (200,))
for r in range(1,201):
    X_r = np.dot(np.dot(U[:,:r] ,np.diag(S[:r])) ,Vt[:r,:])
    error[r-1] = np.linalg.norm(train_data - X_r, ord='fro')
plt.plot(error)
plt.xlabel('r')
plt.ylabel('Rank-r approximation error')
```

Predictive modeling using Logistic Regression on low rank approximation

```python
def r_dim_features(X,train_X,r):
    U,S,Vt = np.linalg.svd(train_X)
    Vtrt = np.transpose(Vt[:r,:])
    return np.dot(X,Vtrt)

r = 10
F = r_dim_features(train_data,train_data,r)
Ft = r_dim_features(test_data,train_data,r)
print train_data.shape,F.shape
print test_data.shape,Ft.shape
```

```python
from sklearn.linear_model import LogisticRegression

r = 10
F = r_dim_features(train_data,train_data,r)
Ft = r_dim_features(test_data,train_data,r)

y_train = train_labels
X_train = F
y_test = test_labels
X_test = Ft

lrm = LogisticRegression(penalty='l2', multi_class='ovr')

lrm.fit(X_train,y_train)
print 'Accuracy when r = 10: ' + str(lrm.score(X_test,y_test))

accuracy = np.ndarray(shape = (200,))
for r in range(1,201):
    F = r_dim_features(train_data,train_data,r)
    Ft = r_dim_features(test_data,train_data,r)
    lrm = LogisticRegression(penalty='l2', multi_class='ovr')
    lrm.fit(F,y_train)
    accuracy[r-1] = lrm.score(Ft,y_test)
plt.plot(accuracy)
plt.xlabel('r')
plt.ylabel('Accuracy')
```
